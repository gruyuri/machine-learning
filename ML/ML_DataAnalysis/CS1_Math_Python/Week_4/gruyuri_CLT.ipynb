Импорт необходимых библиотек

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Для примера выбрал бета-распределение

```python
alpha = 1.0
beta = 4.0
beta_rv = sts.beta(alpha, beta)
sample = beta_rv.rvs(1000)
```

Отобразим гистограмму и функцию плотности распределения

```python
x = np.linspace(0,1,100)
pdf = beta_rv.pdf(x)
plt.plot(x, pdf, label=u'Теоретическая бета-функция')

plt.hist(sample, normed=True)
plt.ylabel(u'Часть примеров из выборки')
plt.xlabel('$x$')
```

Для N=10 сгенерировать 1000 выборок и построить гисторамму распределений их выборочного среднего

```python
N=10
# Массив значений средних по сгенерированным выборкам
sample_val = np.zeros(N)
for i in range(N):
    sub_sample = beta_rv.rvs(1000)
    sample_val[i] = sub_sample.mean()
    
plt.hist(sample_val, normed=True)
plt.ylabel(u'Среднее по выборке для N=10')
plt.xlabel('$x$')
```

Аналогично - для N = 50

```python
N=50
# Массив значений средних по сгенерированным выборкам
sample_val = np.zeros(N)
for i in range(N):
    sub_sample = beta_rv.rvs(1000)
    sample_val[i] = sub_sample.mean()
    
plt.hist(sample_val, normed=True)
plt.ylabel(u'Среднее по выборке для N=50')
plt.xlabel('$x$')
```

И для выборки объема N=100

```python
N=100
# Массив значений средних по сгенерированным выборкам
sample_val = np.zeros(N)
for i in range(N):
    sub_sample = beta_rv.rvs(1000)
    sample_val[i] = sub_sample.mean()
    
plt.hist(sample_val, normed=True)
plt.ylabel(u'Среднее по выборке для N=100')
plt.xlabel('$x$')
```

Вычислим теоретические средние и дисперсии для выборок указанных объемов

```python
# Теоретические статистики для бета-распределения
ex_beta = alpha / (alpha + beta)
dx_beta = alpha*beta / ((alpha + beta)**2 * (alpha + beta + 1))

# Статистики для выборок согласно центральной предельной теореме
import math
ex10 = ex_beta
dx10 = dx_beta / 10.0
sigma10 = math.sqrt(dx10)

ex50 = ex_beta
dx50 = dx_beta / 50.0
sigma50 = math.sqrt(dx50)

ex100 = ex_beta
dx100 = dx_beta / 100.0
sigma100 = math.sqrt(dx100)
```

Гистограмма и теоретическое нормальное распределение для N = 10

```python
N=10
norm_rv = sts.norm(loc=ex10, scale=sigma10)

x = np.linspace(0,1,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label=u'Нормальное согласно ЦПТ', color = 'r')

sample_val = np.zeros(N)
for i in range(N):
    sub_sample = beta_rv.rvs(1000)
    sample_val[i] = sub_sample.mean()
    
plt.hist(sample_val, color = 'b', normed = True)
plt.ylabel(u'Среднее по выборке для N='+str(N))
plt.xlabel('$x$')
```

```python
N=50
norm_rv = sts.norm(loc=ex50, scale=sigma50)

x = np.linspace(0,1,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label=u'Нормальное согласно ЦПТ', color = 'r')

sample_val = np.zeros(N)
for i in range(N):
    sub_sample = beta_rv.rvs(1000)
    sample_val[i] = sub_sample.mean()
    
plt.hist(sample_val, color = 'b', normed = True)
plt.ylabel(u'Среднее по выборке для N='+str(N))
plt.xlabel('$x$')
```

```python
N=100
norm_rv = sts.norm(loc=ex100, scale=sigma100)

x = np.linspace(0,1,100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label=u'Нормальное согласно ЦПТ', color = 'r')

sample_val = np.zeros(N)
for i in range(N):
    sub_sample = beta_rv.rvs(1000)
    sample_val[i] = sub_sample.mean()
    
plt.hist(sample_val, color = 'b', normed = True)
plt.ylabel(u'Среднее по выборке для N='+str(N))
plt.xlabel('$x$')
```

```python
# Еще больше увеличим N
N=1000
ex_N = ex_beta
dx_N = dx_beta / N
sigma_N = math.sqrt(dx_N)
norm_rv = sts.norm(loc=ex_N, scale=sigma_N)

x = np.linspace(0,1,100)
pdf_N = norm_rv.pdf(x)

plt.plot(x, pdf_N, label=u'Нормальное согласно ЦПТ', color = 'r')

sample_val = np.zeros(N)
for i in range(N):
    sub_sample = beta_rv.rvs(1000)
    sample_val[i] = sub_sample.mean()
    
plt.hist(sample_val, color = 'b', normed = True)
plt.ylabel(u'Среднее по выборке для N='+str(N))
plt.xlabel('$x$')
```

# Выводы

Выбранное распределение случайной величины (бета-распределение с параметрами 1 и 4) имеет смещенную функцию плотности распределения с "толстым" правым хвостом.
Постепенно увеличивая число выборок по N (10, 50, 100, 1000), только в последнем случае мы видим совпадение гистограммы и графика плотности нормального распределения с параметрами согласно Центральной Предельной Теореме. Для всех значений N до 100 включительно на графиках представлено только совпадение по среднему (математическому ожиданию), тогда как мода существенно отличается (в данных по выборке она больше по сравнению с теоретическим нормальным распределением).
Тем самым подтверждается утверждение теоремы, что с ростом N кривые совпадают.
